```python
import snscrape.modules.twitter as sntwitter
```

```python
import pandas as pd
```

```python
tweets = []
limit = 100
user= 'from:APompliano OR from:TheCryptoLark OR from:alexkehr OR from:elonmusk OR from:MrBTCFlorida OR from:BitcoinTina OR from:Tyler OR from:VitalikButerin OR from:TylerSCrypto OR from:TylerSCrypto OR from:Bitcoin OR from:FedericoGalavis OR from:BTCTN OR from:BitcoinMagazine OR from:IvanOnTech OR from:BitcoinBob OR from:LayahHeilpern OR from:DanHeld OR from:BTC_Archive OR from:gaborgurbacs OR from:CryptoNewswire'
for k in range (6,9):
    for i in range (2,32):
        if i < 8: 
            var = "0" + str(i)
        else:
            var = str(i)
        since_date = "2021-0"+str(k)+"-"+ var
        for j,tweet in enumerate(sntwitter.TwitterSearchScraper(' (bitcoin OR btc OR BTC) (min_faves:20) lang:en until:{}'.format(since_date), top = True).get_items()):
            if j == limit:
                break
            else:
                tweets.append([tweet.date, tweet.rawContent, tweet.user.username])
tweets_df = pd.DataFrame(tweets, columns=['Date', 'Text', 'user'])
```

```python
# Remove duplicate tweets
init_count = len(tweets_df)
tweet_unique = tweets_df.Text.drop_duplicates(keep=False)
records_with_same_tweet = tweets_df[True ^ tweets_df.Text.isin(tweet_unique)]
records_with_same_tweet[['Text']]
tweets_df = tweets_df[True ^ tweets_df.Text.isin(records_with_same_tweet.Text)]
curr_count = len(tweets_df)
print(f"{init_count - curr_count} duplicate rows dropped!\nCurrent row count: {curr_count}\n\n")
```
